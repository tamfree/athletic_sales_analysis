```python
# Import Libraries and Dependencies
import pandas as pd
#import datetime as dt
```

```python
#CONSTANTS
INVOICE_DATE_COL = "invoice_date"
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020_df = pd.read_csv("Resources/athletic_sales_2020.csv")
sales_2021_df = pd.read_csv("Resources/athletic_sales_2021.csv")
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
sales_2020_21_df = pd.concat([sales_2021_df, sales_2020_df], axis = "rows", join = "inner")
sales_2020_21_df
```

```python
# Check if any values are null.
sales_2020_21_df.isnull().sum()
```

```python
# Check the data type of each column
sales_2020_21_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
sales_2020_21_df[INVOICE_DATE_COL] = pd.to_datetime(sales_2020_21_df[INVOICE_DATE_COL])
sales_2020_21_df
```

```python
# Confirm that the "invoice_date" data type has been changed.
sales_2020_21_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
UNITS_SOLD_COL = "units_sold"
TOTAL_PROD_SOLD_COL = "Total_Products_Sold"
REGION_COL = "region"
STATE_COL = "state"
CITY_COL = "city"
num_prods_df = sales_2020_21_df.groupby([REGION_COL,STATE_COL, CITY_COL])[[UNITS_SOLD_COL]].sum()
num_prods_df = num_prods_df.sort_values(by=UNITS_SOLD_COL, ascending=False)
num_prods_df = num_prods_df.rename(columns={UNITS_SOLD_COL:TOTAL_PROD_SOLD_COL})

# Show the top 5 results.
num_prods_df.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
tot_prod_sold2_df = pd.pivot_table(sales_2020_21_df, \
                                   values= UNITS_SOLD_COL, \
                                    index=[REGION_COL, STATE_COL, CITY_COL],
                                    aggfunc= 'sum',sort = True)
tot_prod_sold2_sorted_df = tot_prod_sold2_df.sort_values(by=UNITS_SOLD_COL, ascending=False)

# Rename the "units_sold" column to "Total_Products_Sold"
tot_prod_sold2_sorted_df = tot_prod_sold2_sorted_df.rename(columns={UNITS_SOLD_COL: TOTAL_PROD_SOLD_COL})

# Show the top 5 results.
tot_prod_sold2_sorted_df.head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
#Additional Column Name Constants
PRODUCT_COL = "product"
WOMENS_FOOTWEAR_UNITS_SOLD_COL = "Womens_Footwear_Units_Sold"
```

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_footwear_sales_df = sales_2020_21_df[sales_2020_21_df[PRODUCT_COL] == "Women's Athletic Footwear"]
womens_footwear_sales_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_footwear_sales_df2 = womens_footwear_sales_df.groupby([RETAILER_COL, REGION_COL, STATE_COL, CITY_COL])[[UNITS_SOLD_COL]].sum()
womens_footwear_sales_df2 = womens_footwear_sales_df2.sort_values(by= UNITS_SOLD_COL, ascending=False)

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sales_df2 = womens_footwear_sales_df2.rename(columns={UNITS_SOLD_COL: WOMENS_FOOTWEAR_UNITS_SOLD_COL})

# Show the top 5 results.
womens_footwear_sales_df2.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
womens_footwear_sales_df3 = pd.pivot_table(womens_footwear_sales_df, index=[RETAILER_COL, STATE_COL, CITY_COL], values=UNITS_SOLD_COL, aggfunc= "sum" )
womens_footwear_sales_df3 = womens_footwear_sales_df3.sort_values(by=UNITS_SOLD_COL, ascending=False)

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
womens_footwear_sales_df3 = womens_footwear_sales_df3.rename(columns={UNITS_SOLD_COL: WOMENS_FOOTWEAR_UNITS_SOLD_COL})

# Show the top 5 results.
womens_footwear_sales_df3.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
sales_times_series_df = pd.pivot_table(womens_footwear_sales_df, index = INVOICE_DATE_COL, values = TOT_SALES_COL)

# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
sales_times_series_df
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in descending order on "Total Sales".
sales_daily_df = sales_times_series_df.resample('D').sum().sort_values(by= TOT_SALES_COL, ascending=False)

sales_daily_df
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in descending order on "Total Sales".
sales_weekly_df = sales_times_series_df.resample('W').sum().sort_values(by= TOT_SALES_COL, ascending=False)
sales_weekly_df = sales_weekly_df.rename(columns={TOT_SALES_COL: "Weekly Total Sales", INVOICE_DATE_COL: "Invoice Week"}) 
week_w_most_sales = sales_weekly_df.iloc[0:1].copy()
week_w_most_sales
```
